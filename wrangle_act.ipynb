```python
predicted_breeds_url = 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'
response = requests.get(predicted_breeds_url)
with open('image_predictions.tsv', 'wb') as f:
    f.write(response.content)
# Import predicted breeds file
df2_original = pd.read_csv("image_predictions.tsv", sep='\t')
```

```python
#Converting data from tweepy to a data frame
df3_original = pd.DataFrame(columns=['tweet_id', 'retweet_account', 'favorite_count'])
with open('tweet_json.txt') as f:
    for line in f:
        status = json.loads(line)
        tweet_id = status['id_str']
        retweet_count = status['retweet_count']
        favorite_count = status['favorite_count']
        df3_original = df3_original.append(pd.DataFrame([[tweet_id, retweet_count, favorite_count]],
                            columns=['tweet_id', 'retweet_count', 'favorite_count']))

df3_original = df3_original.reset_index(drop=True)
```

#Define: extracting the numerator from the ratings. Convert rating_numerator to a float so we can compare to the new extracted numerator values to find discrepancies


Code

```python
df['Numerator']=df.text.str.extract('(\d+[\.]?[\d+]*/)')
df['Numerator']=df['Numerator'].str[0:-1].astype(float)
df["rating_numerator"] = df["rating_numerator"].astype(float)
```

```python
df[df.Numerator!=df.rating_numerator]
```

Define: extracting the Denominator from the ratings, #convert str to int


Code:


```python
df['Denominator']=df.text.str.extract('/([0-9]+)')
df["Denominator"] = pd.to_numeric(df["Denominator"])
```

Define: Combine stages into one column

```python
df['Doggo'].replace(to_replace=np.NaN, value="", inplace=True)
```

```python
# Then combine stage columns.
df['stage'] = df.Doggo + df.Floof + df.Pupper + df.Puppo +df.Blep

```

```python
# Then format entries with multiple dog stages which appeared like doggopupper.

df.loc[df.stage == 'doggopupper', 'stage'] = 'doggo,pupper' 
df.loc[df.stage == 'doggopuppo', 'stage'] = 'doggo,puppo' 
df.loc[df.stage == 'doggofloof', 'stage'] = 'doggo,floof'
df.loc[df.stage == 'pupperblep', 'stage'] = 'pupper,blep'
df.loc[df.stage == 'floofpupper', 'stage'] = 'pupper,floof'
```

#Code: change the cell to NaN if the p1_dog, p2_dog, p3_dog is False


```python
for i in df.p1.index:
    if df.p1_dog[i]==False:
        df.p1[i]=np.NaN
for i in df.p2.index:
    if df.p2_dog[i]==False:
        df.p2[i]=np.NaN   
for i in df.p3.index:
    if df.p3_dog[i]==False:
        df.p3[i]=np.NaN
```

Define: If the p1, p2, or p3 values are null then dog name will be one which is not null

Code:

```python
df['Dog']=""
```

```python
for i in df.index:
    if df.p1[i]!=np.NaN:
        df.Dog[i]=df.p1[i]
    elif df.p2[i]!=np.NaN:
        df.Dog[i]=df.p2[i]
    elif df.p3[i]!=np.NaN:
        df.Dog[i]=df.p3[i]
    else:
        df.Dog[i]=np.NaN
    
```

```python
df.Dog=df.Dog.str.capitalize()
```
